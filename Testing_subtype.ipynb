Loading the VGG19 model

```python
#Loading vgc19 model
from tensorflow.keras.applications import VGG19 # VGG19
vgg_conv = VGG19(weights='imagenet',include_top=False,input_shape=(224, 224, 3))
```

Preprocess the image to fed into the model

```python
def read_image(file_path):
   print('[INFO] loading and preprocessing image…') 
   image = load_img(file_path, target_size=(224, 224)) 
   image = img_to_array(image) 
   image = np.expand_dims(image, axis=0)
   image /= 255. 
   return image
```

function to extract the features and predict the class

```python
def test_single_image(path):
  types = ['Basophil' ,'Eosinophil','Lymphocyte','Monocyte','Neutrophil']
  images = read_image(path)
  time.sleep(.5)
  bt_prediction = vgg_conv.predict(images)
  model = keras.models.load_model('/content/drive/MyDrive/WBC/model_classification.h5')
  preds = model.predict(bt_prediction)
  for idx, type, x in zip(range(0,6), types , preds[0]):
   print('ID: {}, Label: {} {}%'.format(idx, type, round(x*100,2) ))
  
  return
```

```python
path = '/content/recons_img.png'
test_single_image(path)
```
